```python
# 首先确保 dateutil 已安装（如果还没有）
# !conda install -y python-dateutil

# 重启 kernel 后运行以下代码
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
import warnings
warnings.filterwarnings('ignore')

# 检查系统中实际可用的 CJK 字体
print("Available CJK fonts:")
cjk_fonts = []
for f in fm.fontManager.ttflist:
    if any(keyword in f.name.lower() for keyword in ['noto', 'sans', 'serif', 'cjk']):
        if f.name not in cjk_fonts:
            cjk_fonts.append(f.name)
            if len(cjk_fonts) <= 10:  # 只显示前10个
                print(f"  - {f.name}")

# 正确的 matplotlib 配置（使用有效的 rc 参数）
plt.rcParams['font.sans-serif'] = [
    'Noto Sans CJK JP',    # 日文思源黑体（支持中文字符）
    'Noto Serif CJK JP',   # 日文思源宋体
    'DejaVu Sans',         # 英文字体备选
    'Arial Unicode MS'     # 如果存在的话
]
plt.rcParams['axes.unicode_minus'] = False  # 正确显示负号
plt.rcParams['figure.figsize'] = [14, 10]   # 图形尺寸
plt.rcParams['figure.dpi'] = 150            # DPI 设置

# 注意：retina 格式应该在 Jupyter 中通过 magic 命令设置，而不是 rcParams
# 在 Jupyter notebook 中使用：%config InlineBackend.figure_format = 'retina'

# 测试中文显示（但根据项目规范，正式图表应使用英文）
fig, ax = plt.subplots()
ax.plot([1, 2, 3], [1, 4, 2])
ax.set_title('测试中文显示')
ax.set_xlabel('X轴')
ax.set_ylabel('Y轴')
plt.show()

# 符合项目规范的英文版本
fig2, ax2 = plt.subplots()
ax2.plot([1, 2, 3], [1, 4, 2])
ax2.set_title('Test Plot (English)')
ax2.set_xlabel('X-axis')
ax2.set_ylabel('Y-axis')
plt.show()
```
