图像的物理尺寸（inches）保持不变
DPI值会翻倍（例如从72变成144）
坐标轴的数据范围完全相同
结论：
这不是一个bug，而是预期的行为。retina 模式是为了在高分辨率显示器上提供更清晰的图像显示。如果您不希望图像看起来更大，可以：

保持默认设置：不使用 %config InlineBackend.figure_format = 'retina'
调整图像尺寸：使用 plt.figure(figsize=(width, height)) 来控制图像的物理尺寸
理解这是显示优化：实际上您的数据和坐标轴范围完全没有改变
所以，您观察到的"轴延长1.8倍"实际上是由于高DPI渲染造成的视觉错觉，而不是真正的坐标轴缩放。

线型就这4种
| Symbol | Line type | Example                                  |
| ------ | --------- | ---------------------------------------- |
| `-`    | solid     | `plt.plot(x, y, 'r-')`  → red solid      |
| `--`   | dashed    | `plt.plot(x, y, 'g--')` → green dashed   |
| `:`    | dotted    | `plt.plot(x, y, 'b:')`  → blue dotted    |
| `-.`   | dash-dot  | `plt.plot(x, y, 'k-.')` → black dash-dot |


```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 2*np.pi, 100)
y = np.sin(x)

plt.plot(x, y,     color='firebrick', linestyle='solid',   label='solid')
plt.plot(x, y+0.2, color='royalblue', linestyle='dashed',  label='dashed')
plt.plot(x, y+0.4, color='green'    , linestyle='dotted',  label='dotted')
plt.plot(x, y+0.6, color='black'    , linestyle='dashdot', label='dash-dot')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 2*np.pi, 100)
y = np.sin(x)
#  注意, 有keyward 必须在后面, 位置参数必须在前面 (x,y 必须放在前面, 无keyword 的 "-" must be placed before keyword arguments)
plt.plot(x,     y,   '-', color='firebrick', label='solid') # The main difference from the above is that there is no linestyle keyword.
plt.plot(x, y+0.2,  '--', color='royalblue', label='dashed')
plt.plot(x, y+0.4,   ':', color='green',     label='dotted')
plt.plot(x, y+0.6,  '-.', color='black',     label='dash-dot')
plt.legend()
plt.show()
```

5.用饼图的面积及颜色展示一组4维数据

```python
# 创建一个伪随机数生成器的实例，并设置种子为0。
rng = np.random.RandomState(0) # seed the random number generator, as 0 is constant, the same numbers will be generated every time.

x      = rng.randn(100) # 生成一个包含100个随机数的一维数组，这些数服从标准正态分布（均值为0，方差为1）。
y      = rng.randn(100) 
colors = rng.rand(100)  # 生成一个包含100个随机数的一维数组，这些数服从[0,1)上的均匀分布

# plt.scatter()函数接收到这100个数值后，会使用cmap='viridis'这个颜色映射表（colormap）将这些数值转换成具体的颜色。

sizes = 500 * rng.rand(100) # 将每个数乘以500。这些结果值将决定每个点的大小。
#print(sizes)


# 调用scatter函数来绘制散点图。
# x: 指定所有点的x坐标。
# y: 指定所有点的y坐标。
# c=colors: 将colors数组中的值传递给参数c，用于指定每个点的颜色。matplotlib会根据cmap将这些数值映射成具体颜色。
# s=sizes: 将sizes数组中的值传递给参数s，用于指定每个点的面积大小。
# alpha=0.3: 设置所有点的透明度为30%（0.3），这样可以更容易地看到重叠的点。
# cmap='viridis': 指定使用的颜色映射表（colormap）为'viridis'。
# 它会把colors数组中的数值（从0到1）映射为从紫色到黄色的渐变色。
plt.scatter(x, y, c=colors, s=sizes, alpha=0.3, cmap='viridis') # <<==============

# 在图表的旁边添加一个颜色条（colorbar）。
# 这个颜色条会显示数值（来自colors数组）与实际颜色之间的对应关系。
plt.colorbar() ;
# 同时，这个函数会创建并返回一个 Colorbar 对象。这个对象包含了颜色条的所有信息和属性。

# 分号 (;) 可以用来在同一行中分隔多条语句。
# 然而，在Jupyter和IPython这样的环境中，它还有一个特殊的、更常见的用途：抑制当前行代码的输出:
# matplotliobect dadad something
# 
# 在进行数据可视化时，我们通常只关心生成的图像，而不关心绘图函数返回的对象本身。
# 因此，在Jupyter环境中，为了保持输出的整洁，在绘图命令（尤其是单元格的最后一行）的末尾加上分号是一种非常普遍且推荐的做法。

# （可选，但在脚本中通常是必需的）显示生成的图形。
# plt.show()
```

fmt (format)的 “省略线型” 技巧：

fmt : optional
A format string, e.g. 'ro' for red circles. 

section for a full description of the format strings.
Format strings are just an abbreviation for quickly setting
basic line properties. All of these and more can also be
controlled by keyword arguments.

| 写法                     | 效果              | 等价全关键字             |
| ---------------------- | --------------- | ------------------ |
| `fmt='.'`              | **仅圆点**，**无线段** | `linestyle='None'` |

2. 定义边界：N 个颜色需要 N+1 个边界。
bounds = np.arange(N_colors + 1) - 0.5 
例如，对于 N=6，bounds 将是 [-0.5, 0.5, 1.5, 2.5, 3.5, 4.5, 5.5]，这能完美地将整数数据 0-5 分段

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.colors as colors # 导入 colors 模块

# 假设 I 是您的图像数据
I = np.sin(x) * np.cos(x[:, np.newaxis])

# -------------------------- 新方法 --------------------------
N_colors = 6 # 您想要的离散颜色数量

# 1. 获取基础颜色映射
cmap_base = plt.colormaps['Blues']

# 2. 定义边界：创建 N+1 个边界，从 -1.0 到 1.0 均匀分布
#    np.linspace(-1, 1, N_colors + 1) 会生成 7 个边界点：
#    [-1.0, -0.666..., -0.333..., 0.0, 0.333..., 0.666..., 1.0]
bounds = np.linspace(-1.0, 1.0, N_colors + 1)

# 3. 创建归一化对象
#    它将数据 I 映射到 N 个分段，并使用 cmap_base 中的 N 种颜色
norm = colors.BoundaryNorm(bounds, cmap_base.N) 

# 4. 绘制图像
plt.imshow(I, cmap=cmap_base, norm=norm)

# 可选：添加颜色条来显示分段
##plt.colorbar(ticks=np.arange(N_colors)) 
plt.colorbar() 
# plt.clim(-1, 1); # 限制颜色条的范围 (-1, 1) 本例中不需要
#plt.show()
```

grid[:-1, 0] 是 Matplotlib 的 GridSpec 切片语法，含义如下：

grid	一个 plt.GridSpec(4, 4) 对象， 

行索引 0→3，列索引 0→3  
[:-1, 0]	行切片：[:-1] = 从第 0 行到倒数第 2 行（即 0,1,2）

列切片：[0] = 仅第 0 列  
结果：占据第 0 列、第 0→2 行 的子区域（左列，不含最后一行）

```python
# 生成二维正态样本（均值 [0,0]，协方差 [[1,1],[1,2]]，3000 点）
mean = [0, 0]
cov = [[1, 1], [1, 2]]
x, y = np.random.multivariate_normal(mean, cov, 3000).T

# 创建 6×6 英寸画布
fig = plt.figure(figsize=(6, 6))

# 4×4 网格，横向/纵向间距 0.2
grid = plt.GridSpec(4, 4, hspace=0.2, wspace=0.2)

# 主散点图：占据右上方 3×3 网格
main_ax = fig.add_subplot(grid[:-1, 1:])   # 行 0-2，列 1-3 , 左上角为 [0 , 0] 位置

# 左侧 Y 直方图：占据左列 0，行 0-2（与主图共享 Y 轴）
y_hist = fig.add_subplot(grid[:-1, 0], xticklabels=[], sharey=main_ax)

# 底部 X 直方图：占据底行 3，列 1-3（与主图共享 X 轴）
x_hist = fig.add_subplot(grid[-1, 1:], yticklabels=[], sharex=main_ax)

# 主图：散点（小点 + 半透明）
main_ax.scatter(x, y, s=3, alpha=0.2)

# 底部直方图：垂直方向，40 箱，填充式
x_hist.hist(x, 40, histtype='stepfilled', orientation='vertical')
x_hist.invert_yaxis()   # 让直方图向下生长（不遮挡主图） <<<======================================= 注意  倒置 !

# 左侧直方图：水平方向，40 箱，填充式
y_hist.hist(y, 40, histtype='stepfilled', orientation='horizontal')
y_hist.invert_xaxis()   # 让直方图向左生长（不遮挡主图） <<<==============================  注意  倒置 ! 朝向左

plt.show()
```
