结论：
这不是一个bug，而是预期的行为。retina 模式是为了在高分辨率显示器上提供更清晰的图像显示。如果您不希望图像看起来更大，可以：

保持默认设置：不使用 %config InlineBackend.figure_format = 'retina'
调整图像尺寸：使用 plt.figure(figsize=(width, height)) 来控制图像的物理尺寸
理解这是显示优化：实际上您的数据和坐标轴范围完全没有改变
所以，您观察到的"轴延长1.8倍"实际上是由于高DPI渲染造成的视觉错觉，而不是真正的坐标轴缩放。

线型就这4种
| Symbol | Line type | Example                                  |
| ------ | --------- | ---------------------------------------- |
| `-`    | solid     | `plt.plot(x, y, 'r-')`  → red solid      |
| `--`   | dashed    | `plt.plot(x, y, 'g--')` → green dashed   |
| `:`    | dotted    | `plt.plot(x, y, 'b:')`  → blue dotted    |
| `-.`   | dash-dot  | `plt.plot(x, y, 'k-.')` → black dash-dot |


```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 2*np.pi, 100)
y = np.sin(x)

plt.plot(x, y,     color='firebrick', linestyle='solid',   label='solid')
plt.plot(x, y+0.2, color='royalblue', linestyle='dashed',  label='dashed')
plt.plot(x, y+0.4, color='green'    , linestyle='dotted',  label='dotted')
plt.plot(x, y+0.6, color='black'    , linestyle='dashdot', label='dash-dot')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 2*np.pi, 100)
y = np.sin(x)
#  注意, 有keyward 必须在后面, 位置参数必须在前面 (x,y 必须放在前面, 无keyword 的 "-" must be placed before keyword arguments)
plt.plot(x,     y,   '-', color='firebrick', label='solid') # The main difference from the above is that there is no linestyle keyword.
plt.plot(x, y+0.2,  '--', color='royalblue', label='dashed')
plt.plot(x, y+0.4,   ':', color='green',     label='dotted')
plt.plot(x, y+0.6,  '-.', color='black',     label='dash-dot')
plt.legend()
plt.show()
```

5.用饼图的面积及颜色展示一组4维数据

```python
# 创建一个伪随机数生成器的实例，并设置种子为0。
rng = np.random.RandomState(0) # seed the random number generator, as 0 is constant, the same numbers will be generated every time.

x      = rng.randn(100) # 生成一个包含100个随机数的一维数组，这些数服从标准正态分布（均值为0，方差为1）。
y      = rng.randn(100) 
colors = rng.rand(100)  # 生成一个包含100个随机数的一维数组，这些数服从[0,1)上的均匀分布

# plt.scatter()函数接收到这100个数值后，会使用cmap='viridis'这个颜色映射表（colormap）将这些数值转换成具体的颜色。

sizes = 500 * rng.rand(100) # 将每个数乘以500。这些结果值将决定每个点的大小。
#print(sizes)


# 调用scatter函数来绘制散点图。
# x: 指定所有点的x坐标。
# y: 指定所有点的y坐标。
# c=colors: 将colors数组中的值传递给参数c，用于指定每个点的颜色。matplotlib会根据cmap将这些数值映射成具体颜色。
# s=sizes: 将sizes数组中的值传递给参数s，用于指定每个点的面积大小。
# alpha=0.3: 设置所有点的透明度为30%（0.3），这样可以更容易地看到重叠的点。
# cmap='viridis': 指定使用的颜色映射表（colormap）为'viridis'。
# 它会把colors数组中的数值（从0到1）映射为从紫色到黄色的渐变色。
plt.scatter(x, y, c=colors, s=sizes, alpha=0.3, cmap='viridis') # <<==============

# 在图表的旁边添加一个颜色条（colorbar）。
# 这个颜色条会显示数值（来自colors数组）与实际颜色之间的对应关系。
plt.colorbar() ;
# 同时，这个函数会创建并返回一个 Colorbar 对象。这个对象包含了颜色条的所有信息和属性。

# 分号 (;) 可以用来在同一行中分隔多条语句。
# 然而，在Jupyter和IPython这样的环境中，它还有一个特殊的、更常见的用途：抑制当前行代码的输出:
# matplotliobect dadad something
# 
# 在进行数据可视化时，我们通常只关心生成的图像，而不关心绘图函数返回的对象本身。
# 因此，在Jupyter环境中，为了保持输出的整洁，在绘图命令（尤其是单元格的最后一行）的末尾加上分号是一种非常普遍且推荐的做法。

# （可选，但在脚本中通常是必需的）显示生成的图形。
# plt.show()
```

section for a full description of the format strings.

2. 定义边界：N 个颜色需要 N+1 个边界。
bounds = np.arange(N_colors + 1) - 0.5 
例如，对于 N=6，bounds 将是 [-0.5, 0.5, 1.5, 2.5, 3.5, 4.5, 5.5]，这能完美地将整数数据 0-5 分段

```python
# =========================
# 详细注释版：散点图 + 边缘直方图布局示例
# =========================
# 生成二维高斯（正态）分布样本
# mean 表示均值向量 [x 的均值, y 的均值]
# 这里两个维度的均值都为 0，所以样本整体会围绕原点附近分布
mean = [0, 0]

# cov 是 2x2 协方差矩阵：
# [[var(x), cov(x,y)],
#  [cov(y,x), var(y)]]
# 其中：
# - x 的方差是 1，说明 x 方向有一定离散程度
# - y 的方差是 2，说明 y 比 x 更分散
# - x 和 y 的协方差是 1，说明二者正相关：x 越大时，y 往往也越大
cov = [[1, 1], [1, 2]]

# 从上述二维正态分布中随机采样 3000 个点
# 返回结果形状大致是 (3000, 2)，每一行是一个点 [x_i, y_i]
# .T 转置后变成 (2, 3000)，这样就可以直接拆成 x 和 y 两个一维数组
x, y = np.random.multivariate_normal(mean, cov, 3000).T

# 创建整张图的画布，尺寸为 6x6 英寸
# figsize 控制的是整个 figure 的物理大小，不是后面 4x4 网格的格子数量
fig = plt.figure(figsize=(6, 6))  # height/width 和 GridSpec 的 4x4 划分没有直接对应关系

# 在整张画布上划分一个 4x4 的布局网格
# hspace=0.2 表示子图之间垂直方向的间距
# wspace=0.2 表示子图之间水平方向的间距
# 这个网格只是用于“排版定位”，并不代表数据坐标
grid = plt.GridSpec(4, 4, hspace=0.2, wspace=0.2)

# 创建主散点图坐标轴，放在右上角的 3x3 区域
# grid[:-1, 1:] 的含义：
#  :-1 表示取第 0,1,2 行（不取最后一行）# -1 指 last line， python的规则是从0开始，最后一个数字结束，但不包括最后一个数字 <<<=============================================
#  1:  表示取第 1,2,3 列（不取第 0 列）
# 所以主图会占据：
#   行 0-2，列 1-3

# 注意：GridSpec 的索引原点在左上角，即左上角位置是 [0, 0]                                                                     <<=============================
main_ax = fig.add_subplot(grid[:-1, 1:])

# 创建左侧的 y 方向直方图
# 放在第 0 列、前 3 行，也就是主图左边那一条区域
# sharey=main_ax 表示与主图共享 y 轴刻度范围                                  <<=============================
# 这样左侧直方图和主散点图在 y 方向上是严格对齐的
# xticklabels=[] 表示不显示这个子图的 x 轴刻度标签，避免画面重复和拥挤           <<=============================
y_hist = fig.add_subplot(grid[:-1, 0], xticklabels=[], sharey=main_ax)

# 创建底部的 x 方向直方图
# 放在最后一行、第 1-3 列，也就是主图正下方
# sharex=main_ax 表示与主图共享 x 轴刻度范围                  <<=============================
# 这样底部直方图和主散点图在 x 方向上是严格对齐的
# yticklabels=[] 表示不显示这个子图的 y 轴刻度标签，减少干扰
x_hist = fig.add_subplot(grid[-1, 1:], yticklabels=[], sharex=main_ax)

# 在主图中绘制散点图
# x, y 分别是每个样本点的横纵坐标
# s=3 表示点很小，这样 3000 个点叠在一起时不会太拥挤  <<=============================
# alpha=0.2 表示透明度较低，点重叠越多的区域会显得越深，更容易看出密度
main_ax.scatter(x, y, s=3, alpha=0.2)

# 在底部子图中画 x 的直方图
# 这里统计的是 x 这一维的边缘分布
# 40 表示把数据分成 40 个箱（bin）
# histtype='stepfilled' 表示绘制带填充的轮廓型直方图
# orientation='vertical' 表示柱子沿竖直方向生长（默认也是竖直）
x_hist.hist(x, 40, histtype='stepfilled', orientation='vertical')

# 将底部直方图的 y 轴倒过来                              <<=============================
# 默认情况下，柱子会“向上长”，这会挤向主图区域
# 倒置之后，柱子改为“向下长”，从而与主图分离，视觉上更清晰
x_hist.invert_yaxis()

# 在左侧子图中画 y 的直方图
# 这里统计的是 y 这一维的边缘分布
# orientation='horizontal' 表示柱子沿水平方向生长
# 也就是说，频数会朝左右方向展开，而不是上下方向
y_hist.hist(y, 40, histtype='stepfilled', orientation='horizontal')

# 将左侧直方图的 x 轴倒过来
# 默认情况下，水平直方图会向右生长，这样会顶到主图            <<=============================
# 倒置之后，柱子向左生长，从而放在主图外侧，不遮挡主图内容
y_hist.invert_xaxis()

# 显示整张图
plt.show()
```

```python
# =========================
# 教学版：散点图 + 边缘直方图布局示例
# =========================
# 这段代码的目标是画出一种很常见的统计可视化：
# 1. 中间/右上区域：主散点图，用来看二维变量 (x, y) 的联合分布
# 2. 下方：x 的直方图，用来看 x 单独的分布
# 3. 左侧：y 的直方图，用来看 y 单独的分布
#
# 这种图的好处是：
# - 主图告诉你 x 和 y 是否相关、点云形状如何
# - 两个边缘直方图告诉你每个变量各自的分布情况
# - 三张图共用同一套 x / y 范围，因此可以直接对照观察

# -------------------------
# 第 1 步：生成二维正态分布样本
# -------------------------

# mean 是均值向量，表示二维随机变量的中心位置
# [0, 0] 的意思是：
# - x 的平均值是 0
# - y 的平均值是 0
# 所以整体点云会大致围绕原点附近分布
mean = [0, 0]

# cov 是协方差矩阵，用来描述：
# 1. 每个变量本身有多分散
# 2. 两个变量之间有没有线性相关
#
# 矩阵写法：
# [[var(x),   cov(x,y)],
#  [cov(y,x), var(y)]]
#
# 这里是：
# [[1, 1],
#  [1, 2]]
#
# 可以这样理解：
# - var(x)=1：x 有一定波动
# - var(y)=2：y 的波动比 x 更大
# - cov(x,y)=1：x 和 y 正相关
#   即：x 越大时，y 往往也会偏大
#
# 因此最后画出来的散点云通常不是“圆形”，
# 而是沿某个斜方向拉长的椭圆形
cov = [[1, 1], [1, 2]]

# 从二维正态分布中抽样 3000 个点
# np.random.multivariate_normal(mean, cov, 3000)
# 返回一个形状为 (3000, 2) 的数组：
#   第 0 列是所有点的 x
#   第 1 列是所有点的 y
#
# .T 表示转置，转成 (2, 3000)
# 这样就能很方便地拆成两个一维数组：
#   x: 长度为 3000
#   y: 长度为 3000
x, y = np.random.multivariate_normal(mean, cov, 3000).T

# -------------------------
# 第 2 步：创建整张画布 figure
# -------------------------

# figure 可以理解为“整张纸”或“整块画板”
# figsize=(6, 6) 表示整张图的物理尺寸是 6 英寸 × 6 英寸
#
# 注意：
# 这里的 6×6 只决定整张图有多大，
# 跟后面说的“4×4 网格”不是一回事
#
# 换句话说：
# - figsize 决定“画布大小”
# - GridSpec 决定“怎么切分布局”
fig = plt.figure(figsize=(6, 6))

# -------------------------
# 第 3 步：把整张画布切成 4×4 网格
# -------------------------

# GridSpec(4, 4) 的意思是：
# 把整张 figure 在布局上分成 4 行 4 列，一共 16 个小格子
#
# 这些格子只是“排版格子”，不是数据坐标，不代表 x/y 数值
#
# hspace=0.2：
#   子图之间上下方向的留白
# wspace=0.2：
#   子图之间左右方向的留白
grid = plt.GridSpec(4, 4, hspace=0.2, wspace=0.2)

# 如果把网格编号画出来，大致可以想成：
#
#   列 ->   0     1     2     3
# 行
# 0       [ ]   [ ]   [ ]   [ ]
# 1       [ ]   [ ]   [ ]   [ ]
# 2       [ ]   [ ]   [ ]   [ ]
# 3       [ ]   [ ]   [ ]   [ ]
#
# 并且注意：
# 左上角是 [0, 0]
# 也就是：
# - 行号从上往下增加
# - 列号从左往右增加

# -------------------------
# 第 4 步：放置主散点图 main_ax
# -------------------------

# grid[:-1, 1:] 这个切片是本段代码的关键之一
#
# 先拆开看：
# 1. :-1   表示“从开头到最后一行之前”
#    对 4 行来说，就是取第 0、1、2 行
#
# 2. 1:    表示“从第 1 列到最后”
#    对 4 列来说，就是取第 1、2、3 列
#
# 所以 grid[:-1, 1:] 最终选中的是：
# - 行 0,1,2
# - 列 1,2,3
#
# 也就是说，主图占据右上角一个 3×3 的大区域
main_ax = fig.add_subplot(grid[:-1, 1:])

# 你可以把它想象成：
#
#   [左直方图] [       主散点图区域        ]
#   [左直方图] [       主散点图区域        ]
#   [左直方图] [       主散点图区域        ]
#   [   空/角落 ] [   底部 x 直方图区     ]
#
# 这里主图没有占最左列，也没有占最后一行，
# 因为这些位置要留给两个边缘直方图

# -------------------------
# 第 5 步：放置左侧 y 直方图 y_hist
# -------------------------

# grid[:-1, 0] 表示：
# - 行 0、1、2
# - 列 0
#
# 也就是主图左边那一条竖着的区域
#
# sharey=main_ax 很重要：
# 它表示 y_hist 和主图共用同一个 y 轴范围
#
# 这会带来两个效果：
# 1. 左侧直方图的纵向位置和主图严格对齐
# 2. 主图里 y=1 的高度，和左图里 y=1 的高度是同一条水平线
#
# 这正是“边缘直方图”可读性强的关键
#
# xticklabels=[] 表示不显示 x 轴刻度文字
# 因为左图的主要任务只是辅助展示分布，不需要重复太多标签
y_hist = fig.add_subplot(grid[:-1, 0], xticklabels=[], sharey=main_ax)

# -------------------------
# 第 6 步：放置底部 x 直方图 x_hist
# -------------------------

# grid[-1, 1:] 表示：
# - 最后一行（第 3 行）
# - 第 1、2、3 列
#
# 也就是主图下方那一条横着的区域
#
# sharex=main_ax 同样很重要：
# 它表示 x_hist 和主图共用同一个 x 轴范围
#
# 这会带来两个效果：
# 1. 底部直方图的横向位置和主图严格对齐
# 2. 主图里 x=2 的位置，和下图里 x=2 的位置是同一条竖线
#
# yticklabels=[] 表示不显示 y 轴刻度文字
# 因为这里更关注形状，不强调具体频数标签
x_hist = fig.add_subplot(grid[-1, 1:], yticklabels=[], sharex=main_ax)

# -------------------------
# 第 7 步：绘制主散点图
# -------------------------

# scatter(x, y) 画出 3000 个散点
#
# s=3：
#   每个点的尺寸很小
#   点太大时，3000 个点会互相遮挡得很严重
#
# alpha=0.2：
#   设置透明度
#   单个点比较淡，但很多点重叠的地方会更深
#   这样能更容易看出哪里密度高、哪里密度低
main_ax.scatter(x, y, s=3, alpha=0.2)

# -------------------------
# 第 8 步：绘制底部 x 直方图
# -------------------------

# 这张图只统计 x 的分布，不关心 y
#
# hist(x, 40, ...)
# 表示把 x 的取值范围切成 40 个箱子（bin）
# 然后统计每个区间里落入了多少个样本
#
# orientation='vertical'
# 表示柱子默认朝“上下方向”长
# 即：
# - 横轴是 x 的取值区间
# - 纵轴是频数/计数
#
# histtype='stepfilled'
# 表示画成带填充的轮廓风格
# 看起来通常比默认实心柱状更简洁一些
x_hist.hist(x, 40, histtype='stepfilled', orientation='vertical')

# invert_yaxis() 是第二个关键点
#
# 默认情况下，底部直方图会“向上长”
# 但如果向上长，就会顶到主图，占据主图下边缘附近的视觉空间
#
# 我们更希望它“向下长”，即远离主图
# 所以把 y 轴倒置后：
# - 数值坐标方向被翻转
# - 视觉上柱子会朝下延伸
#
# 这样布局更自然：
# 主图在上，x 的边缘分布贴在下方并向下展开
x_hist.invert_yaxis()

# -------------------------
# 第 9 步：绘制左侧 y 直方图
# -------------------------

# 这张图只统计 y 的分布，不关心 x
#
# orientation='horizontal'
# 表示画“水平直方图”
# 即：
# - 纵轴是 y 的取值区间
# - 横轴是频数/计数
#
# 这样它就能自然地贴在主图左边，作为 y 的边缘分布
y_hist.hist(y, 40, histtype='stepfilled', orientation='horizontal')

# invert_xaxis() 是第三个关键点
#
# 默认情况下，左侧水平直方图会“向右长”
# 但向右长的话，就会冲进主图区域，遮挡主图左边
#
# 所以这里把 x 轴倒置：
# - 视觉上柱子改为“向左长”
# - 这样直方图就朝主图外侧展开，不会挡住主图
#
# 结果就是：
# - 主图右上
# - x 直方图向下展开
# - y 直方图向左展开
# 三者拼接成一个很清晰的组合图
y_hist.invert_xaxis()

# -------------------------
# 第 10 步：显示图像
# -------------------------

# 最后把整张图显示出来
plt.show()
```
